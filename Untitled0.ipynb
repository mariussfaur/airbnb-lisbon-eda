```python
# ============================================
# Airbnb Lisbon — Exploratory Data Analysis
# Author: Marius Faur
# Tools: Python, Pandas, Matplotlib, Seaborn
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
print("Libraries loaded successfully!")
```

```python
# Upload do ficheiro — corre esta célula e faz upload do listings.csv.gz
from google.colab import files
uploaded = files.upload()

# Carregar o dataset
df = pd.read_csv('listings.csv.gz', compression='gzip', low_memory=False)

# Primeira exploração
print(f"Dataset shape: {df.shape}")
print(f"\nColunas disponíveis:\n{df.columns.tolist()}")
print(f"\nPrimeiras linhas:")
df.head()
```

```python
# Selecionar colunas relevantes
cols = ['id', 'neighbourhood_cleansed', 'room_type', 'price',
        'number_of_reviews', 'review_scores_rating', 'accommodates',
        'bedrooms', 'availability_365']

df = df[cols].copy()

# Limpar coluna de preço (remover $ e converter para número)
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# Remover valores nulos e outliers extremos
df = df.dropna(subset=['price', 'neighbourhood_cleansed'])
df = df[df['price'] > 0]
df = df[df['price'] < 1000]  # remover outliers extremos

print(f"Dataset limpo: {df.shape[0]} alojamentos")
print(f"\nValores nulos restantes:\n{df.isnull().sum()}")
df.describe()
```

```python
# Preço médio por bairro (top 15)
avg_price = (df.groupby('neighbourhood_cleansed')['price']
               .mean()
               .sort_values(ascending=False)
               .head(15)
               .reset_index())

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1 — Preço médio por bairro
sns.barplot(data=avg_price, x='price', y='neighbourhood_cleansed',
            palette='Blues_r', ax=axes[0])
axes[0].set_title('Average Price by Neighbourhood (Top 15)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Average Price (€)')
axes[0].set_ylabel('')

# Gráfico 2 — Distribuição por tipo de quarto
room_counts = df['room_type'].value_counts()
axes[1].pie(room_counts, labels=room_counts.index, autopct='%1.1f%%',
            colors=sns.color_palette('Blues', len(room_counts)))
axes[1].set_title('Distribution by Room Type', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('price_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

```python
# Relação entre preço, rating e disponibilidade
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 3 — Preço vs Rating
df_rated = df.dropna(subset=['review_scores_rating'])
axes[0].scatter(df_rated['review_scores_rating'], df_rated['price'],
                alpha=0.3, color='steelblue', s=10)
axes[0].set_title('Price vs Review Rating', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Review Score')
axes[0].set_ylabel('Price (€)')

# Gráfico 4 — Disponibilidade média por bairro (top 10)
avail = (df.groupby('neighbourhood_cleansed')['availability_365']
           .mean()
           .sort_values(ascending=False)
           .head(10)
           .reset_index())

sns.barplot(data=avail, x='availability_365', y='neighbourhood_cleansed',
            palette='Oranges_r', ax=axes[1])
axes[1].set_title('Average Availability (days/year) — Top 10', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Days Available per Year')
axes[1].set_ylabel('')

plt.tight_layout()
plt.savefig('ratings_availability.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Sumário estatístico dos insights principais
print("=" * 50)
print("KEY INSIGHTS — Airbnb Lisbon EDA")
print("=" * 50)

print(f"\nTotal listings analysed: {len(df):,}")
print(f"Average price: €{df['price'].mean():.2f}")
print(f"Median price: €{df['price'].median():.2f}")

print(f"\nMost expensive neighbourhood: {avg_price.iloc[0]['neighbourhood_cleansed']} "
      f"(avg €{avg_price.iloc[0]['price']:.2f})")
print(f"Most affordable neighbourhood: {avg_price.iloc[-1]['neighbourhood_cleansed']} "
      f"(avg €{avg_price.iloc[-1]['price']:.2f})")

print(f"\nMost common room type: {df['room_type'].value_counts().index[0]} "
      f"({df['room_type'].value_counts().iloc[0] / len(df) * 100:.1f}%)")

print(f"\nAverage review score: {df['review_scores_rating'].mean():.2f}/5")
print(f"Average availability: {df['availability_365'].mean():.0f} days/year")
print("\nAnalysis complete!")
```
